```python
import tensorflow as tf
```

```python
from tensorflow.examples.tutorials.mnist import input_data
mnist=input_data.read_data_sets("MNIST_data/",one_hot=True)
```

```python
print(mnist.train.images.shape,mnist.train.labels.shape)
print(mnist.test.images.shape,mnist.test.labels.shape)
print(mnist.validation.images.shape,mnist.validation.labels.shape)
```

```python
sess=tf.InteractiveSession()
#類似Input size,需指定tensor型別
#Keras.Input=tf.placeholder
x=tf.placeholder(tf.float32,[None,784])
W=tf.Variable(tf.zeros([784,10]))
b=tf.Variable(tf.zeros([10]))
y=tf.nn.softmax(tf.matmul(x,W)+b)

#input size 10*1 tensor for evaluting cross entropy
y_=tf.placeholder(tf.float32,[None,10])
#10個dimension的和(axis=1)在一個batch取平均
cross_entropy=tf.reduce_mean(-tf.reduce_sum(y_*tf.log(y),axis=1))

#設定類似keras的compile
train_step=tf.train.GradientDescentOptimizer(0.5).minimize(cross_entropy)
#參數初始化
tf.global_variables_initializer().run()

for i in range(1):
    batch_xs,batch_ys=mnist.train.next_batch(100)
    train_step.run({x:batch_xs,y_:batch_ys})
    
    #output is label
    correct_prediction=tf.equal(tf.argmax(y,1),tf.argmax(y_,1))
    #output is accuracy(float)
    accuracy=tf.reduce_mean(tf.cast(correct_prediction,tf.float32))
    #類似model.predict
    print(accuracy.eval({x:mnist.test.images,y_:mnist.test.labels}))
```

```python
import keras as K
hello=K.models.Sequential()
hello.add(K.layers.Dense(10,activation='softmax',input_shape=(784,)))
hello.compile('SGD','binary_crossentropy',metrics=[K.metrics.categorical_accuracy])
hello.fit(mnist.train.images,mnist.train.labels,
          validation_data=(mnist.validation.images,mnist.validation.labels),epochs=1000,batch_size=100)
```
