```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
sns.set()
sns.set_style("whitegrid")
from scipy.stats import norm
from scipy import stats
import re
from sklearn.preprocessing import StandardScaler
from sklearn.model_selection import train_test_split
```

```python
train_data=pd.read_csv('train.csv',index_col='Id')
test_data=pd.read_csv('test.csv',index_col='Id')

train_y=train_data['SalePrice']
train_x=train_data.iloc[:,:-1]

all_data=pd.concat([train_x,test_data],axis=0)
print('all_data#:',all_data.shape)
print('train_x#:',train_x.shape)
print('test_data#:',test_data.shape)
```

* Remove Outlier

```python
x=train_data['GrLivArea']
y=train_y
plt.scatter(x,y)
plt.xlabel('GrLivArea')
plt.ylabel('SalePrice')
plt.show()

train_data=train_data.loc[train_data['GrLivArea']<4500]
train_x=train_data.iloc[:,:-1]
train_y=train_data['SalePrice']

all_data=pd.concat([train_x,test_data],axis=0)
x=train_data['GrLivArea']
y=train_y
plt.scatter(x,y)
plt.xlabel('GrLivArea')
plt.ylabel('SalePrice')
plt.show()

print('all_data#:',all_data.shape)
print('train_x#:',train_x.shape)
print('test_data#:',test_data.shape)
```

```python
all_data.columns
```

```python
train_y=np.log(train_y)
#all_data['GrLivArea']=np.log(all_data['GrLivArea'])
#all_data["OverallQual-s2"] = all_data["OverallQual"] ** 2
all_data["HouseYear"] = all_data["YrSold"]-all_data['YearBuilt']
del all_data["YearBuilt"]

print('all_data#:',all_data.shape)

sns.distplot(train_y, fit=norm);
fig = plt.figure()
res = stats.probplot(train_y, plot=plt)
plt.show()
```

* fill in missing value

```python
#checking table
pv_table=pd.pivot_table(all_data.fillna('missing'),values=all_data.columns[0],index='SaleType',aggfunc=len)
print(pv_table)
```

```python
all_data.loc[all_data['MSZoning'].isnull(),'MSZoning']=all_data['MSZoning'].mode()[0]
all_data.loc[all_data['Alley'].isnull(),'Alley']='None'
all_data.loc[all_data['Utilities'].isnull(),'Utilities']=all_data['Utilities'].mode()[0]
all_data.loc[all_data['Exterior1st'].isnull(),'Exterior1st']=all_data['Exterior1st'].mode()[0]
all_data.loc[all_data['Exterior2nd'].isnull(),'Exterior2nd']=all_data['Exterior2nd'].mode()[0]
all_data.loc[all_data['MasVnrType'].isnull(),'MasVnrType']=all_data['MasVnrType'].mode()[0]
all_data.loc[all_data['BsmtQual'].isnull(),'BsmtQual']='None'
all_data.loc[all_data['BsmtCond'].isnull(),'BsmtCond']='None'
all_data.loc[all_data['BsmtExposure'].isnull(),'BsmtExposure']='None'
all_data.loc[all_data['BsmtFinType1'].isnull(),'BsmtFinType1']='None'
all_data.loc[all_data['BsmtFinType2'].isnull(),'BsmtFinType2']='None'
all_data.loc[all_data['Electrical'].isnull(),'Electrical']=all_data['Electrical'].mode()[0]
all_data.loc[all_data['KitchenQual'].isnull(),'KitchenQual']=all_data['KitchenQual'].mode()[0]
all_data.loc[all_data['Functional'].isnull(),'Functional']=all_data['Functional'].mode()[0]
all_data.loc[all_data['FireplaceQu'].isnull(),'FireplaceQu']='None'
all_data.loc[all_data['GarageType'].isnull(),'GarageType']='None'
all_data.loc[all_data['GarageFinish'].isnull(),'GarageFinish']='None'
all_data.loc[all_data['GarageQual'].isnull(),'GarageQual']='None'
all_data.loc[all_data['GarageCond'].isnull(),'GarageCond']='None'
all_data.loc[all_data['PoolQC'].isnull(),'PoolQC']='None'
all_data.loc[all_data['Fence'].isnull(),'Fence']='None'
all_data.loc[all_data['MiscFeature'].isnull(),'MiscFeature']='None'
all_data.loc[all_data['SaleType'].isnull(),'SaleType']=all_data['SaleType'].mode()[0]

#numerical columns which contain missing value are filled in its median
all_data=all_data.fillna(all_data.median())

#categorical not in numerical format(only for MSSubClass)
all_data['MSSubClass']=all_data['MSSubClass'].astype(str)
all_data['MoSold']=all_data['MoSold'].astype(str)
all_data['YrSold']=all_data['YrSold'].astype(str)
#all_data['YearBuilt']=all_data['YearBuilt'].astype(str)
all_data['YearRemodAdd']=all_data['YearRemodAdd'].astype(str)
all_data["HouseYear"]=all_data["HouseYear"].astype(str)
```

* get dummy variables

```python
numeric_list=[]
categorical_list=[]
for i in range(all_data.shape[1]):
    if all_data.iloc[:,i].dtype!='O':
        numeric_list.append(all_data.columns[i])
    else:
        categorical_list.append(all_data.columns[i])
print('categorical#:',len(categorical_list))
print('numerical#:',len(numeric_list))

```

```python
numerical_data=all_data[numeric_list]
categorical_data=all_data[categorical_list]

dummy=pd.get_dummies(categorical_data)
print('dummy shape:',dummy.shape)

r = re.compile('.*None')
newlist = list(filter(r.match, dummy.columns))
for i in newlist:
    del dummy[i]
print('None len:',len(newlist))
print('dummy shape after reducing None:',dummy.shape)

```

```python
new_data=pd.concat([numerical_data,dummy],axis=1)
print('numerical_data shape:',numerical_data.shape)
print('new_data shape:',new_data.shape)
```

* split train test data

```python
training_X=new_data[:train_y.shape[0]]
testing_X=new_data[train_y.shape[0]:]

X_train,X_test,y_train,y_test=train_test_split(training_X,train_y,test_size=0.3,random_state=10)

sc=StandardScaler()
X_train=sc.fit_transform(X_train)
X_test=sc.fit_transform(X_test)
testing_X=sc.fit_transform(testing_X)


print(training_X.shape)
print(testing_X.shape)
```

```python
from sklearn.linear_model import ElasticNet
from sklearn.ensemble import GradientBoostingRegressor
from sklearn.linear_model import Lasso

from sklearn.linear_model import ElasticNetCV
from sklearn.model_selection import cross_val_score,cross_val_predict
```

```python
lasso=Lasso(alpha = 0.0006, max_iter = 50000)
lasso.fit(X_train,y_train)
y_train_pred=cross_val_score(lasso,X_train,y_train,scoring='neg_mean_squared_error',cv=10)
y_test_pred=lasso.predict(X_test)

from sklearn.metrics import mean_squared_error
print('RMSE train: %.3f, test: %.3f' % (
        -(score.mean()),
        np.sqrt(mean_squared_error((y_test), (y_test_pred)))))
```

```python
#regr = ElasticNet(random_state=0,alpha=1,l1_ratio=0.01,max_iter=1000)
'''elasticNet = ElasticNetCV(l1_ratio = 1,
                          alphas = 0.006, 
                          max_iter = 50000,
                          cv=5)'''
elasticNet = ElasticNetCV(l1_ratio =0.1,
                          alphas = [0.03], 
                          max_iter = 50000, cv = 10)

elasticNet.fit(X_train,y_train)
y_train_pred=cross_val_predict(elasticNet,X_test,y_test,cv=10)
y_test_pred=elasticNet.predict(X_test)

from sklearn.metrics import mean_squared_error
print('RMSE train: %.3f, test: %.3f' % (
        -(score.mean()),
        np.sqrt(mean_squared_error((y_test), (y_test_pred)))))

#BaseLine
#RMSE train: 0.127, test: 0.134
#RMSE train: 0.087, test: 0.112(power2 of overallQual)
```

```python
coefs = pd.Series(elasticNet.coef_, index = new_data.columns)
print(coefs.sort_values(ascending=False).head(15))
print('eliminating terms:',(coefs==0).sum())
print('total terms:',(coefs==0).count())
```

```python
est = GradientBoostingRegressor(n_estimators=3000, learning_rate=0.05,
                                   max_depth=4, max_features='sqrt',
                                   min_samples_leaf=15, min_samples_split=10, 
                                   loss='huber', random_state =5)
est.fit(X_train, y_train)
#print('training result:',est.score(X_train,y_train))
#print('testing result:',est.score(X_test,y_test))
y_train_pred=cross_val_predict(est,X_test,y_test,cv=10)
y_test_pred=est.predict(X_test)

from sklearn.metrics import mean_squared_error
print('RMSE train: %.3f, test: %.3f' % (
        -(score.mean()),
        np.sqrt(mean_squared_error((y_test), (y_test_pred)))))
#BaseLine
#RMSE train: 0.054, test: 0.111(power2 of overallQual)
```
